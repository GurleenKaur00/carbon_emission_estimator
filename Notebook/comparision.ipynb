```python
import pandas as pd
import matplotlib.pyplot as plt
```

WRITING ALL MODELS PERFORMANCE METRICS

```python
model_scores = [
    {"Model": "Random Forest", "MAE": 15.04, "RMSE": 43.13, "R² Score": 0.9810},
    {"Model": "XGBoost" ,       "MAE": 13.25, "RMSE": 42.45, "R² Score": 0.9816},
    {"Model": "Linear Regression", "MAE": 37.30, "RMSE": 101.83, "R² Score": 0.8942}
]

# Convert to DataFrame
score_df = pd.DataFrame(model_scores)
```

ML MODEL PERFORMANCE COMPARISION

```python
print("\nML Model Performance Comparison:")
print(score_df)
```

PLOTTING COMPARISION

```python
# Plot
plt.figure(figsize=(10, 6))
score_df.set_index("Model")[["MAE", "RMSE"]].plot(kind="bar", figsize=(10, 6), colormap="Set2")
plt.title("Comparison of MAE and RMSE for Different Models")
plt.ylabel("Error")
plt.grid(axis='y')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

R² PLOTTING 

```python
# R² plot separately
plt.figure(figsize=(8, 4))
plt.bar(score_df["Model"], score_df["R² Score"], color='lightcoral')
plt.title("R² Score Comparison")
plt.ylabel("R² Score")
plt.ylim(0.85, 1.0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
